```python
class prompting():
    def __init__(self):
        pass
    def iterate(self, df, prompt_template, type, args, cols, extract_data, i=0):
        results = []
        
        for index, row in list(df.iterrows())[i:]:
            prompt = prompt_template
            while True:
                try:
                    name = row['organization name']
                    website = row['website']
                    print(f"Processing {index}: {name}, {website}")
                    if "generated_description" in df.columns:
                        print(row["generated_description"])
                    
                    for arg in args: 
                        prompt = prompt.replace(f"${arg[0]}", row[arg[1]])
                    result = x_gemini.ask(prompt)
                    
                    print(f"Result: {result}\n")
                    result_vals = extract_data(result)
                    print(result_vals)
                    results.append([name]+result_vals)
                    break 
                
                except Exception as e:
                    print(f"Error processing {index}, {row['organization name']}: {e}")
                    time.sleep(20)  
        if type in df.columns:
            df = df.drop(columns=[type,type+"_conf_interval", type+"_conf_interval_resoning"])
        self.results_df = pd.DataFrame(results, columns=cols)
        df = df.merge(self.results_df, on='organization name', how='left')
        return df

prompting_class = prompting()
```

```python
def extract_data(text):
    text = text.replace("\n"," ").replace("  "," ").replace("*","").replace("Description:","").replace("/10","").replace("_","").replace("#","")
    pattern = r"^(.*?)\s*Confidence Interval: (\d+)\s+Reasoning: (.*)$"
    match = re.search(pattern, text, re.DOTALL)
    return [match.group(x).strip() for x in range(1, 4)] if match else []

args = [["company","organization name"],["website","website"], ["description","description_all"]]
cols = ["organization name", "generated_description", "generated_description_conf_interval", "generated_description_conf_interval_resoning"]
sample_df = prompting_class.iterate(sample_df, generated_description_prompt, "generated_description", args, cols, extract_data)
```

```python
def parse_data(text, pattern):
    matches = re.findall(pattern, text, flags=re.DOTALL)
    vals = []
    for match in matches:
        vals.extend([m.split(":")[1].strip() if ":" in m else m.strip() for m in match if m != ''])
    print(vals)
    return vals


def extract_data(text):
    text = text.replace("\n"," ").replace("  "," ").replace("*","").replace("Description:","").replace("/10","").replace("_","").replace("#","")
    pattern = r"^(.*?)\s*Confidence Interval: (\d+)\s+Reasoning: (.*)$"
    match = re.search(pattern, text, re.DOTALL)
    vals = [match.group(x).strip() for x in range(1, 4)] if match else []
    
    pattern = r'Tasks/Jobs:\s*(.*?)\s*Industry:\s*(.*?)\s*Customers:\s*(.*)'
    vals += parse_data(vals[0],pattern)
    return vals
    
args = [["company","organization name"],["website","website"], ["description","generated_description"]]
cols = ["organization name", "parsed_description", "parsed_description_conf_interval", "parsed_description_conf_interval_resoning", "Tasks/Jobs","Industry","Customers"]
sample_df = prompting_class.iterate(sample_df, parsed_description_prompt, "parsed_description", args, cols, extract_data)

```
