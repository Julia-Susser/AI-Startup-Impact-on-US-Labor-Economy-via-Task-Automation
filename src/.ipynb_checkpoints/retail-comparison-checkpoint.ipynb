```python
filter_prompt = """
On a scale of 1-100, Is the goal of the product of $company ($website) to be used by a task & occupation in the US workforce? 

For each startup, provide
    + a specific rating (1-100) of how likely tool would be to replace part of a human workforce activity. 
    + if it a tool to help someone at work, then lower the rating. If it is automating part of the job independently, then give higher rating.
    + If rating > 50, give ONET job title that the tool is automating.

**EXAMPLES BELOW**
Description: DecoverAI is an advanced research and eDiscovery platform tailored specifically for legal teams. It leverages cutting-edge AI technology to streamline and accelerate the discovery process, enabling lawyers to uncover critical insights and evidence efficiently.
Rating: 98
Reasoning (one sentence): Replacing research task for paralegal
ONET Job: Paralegal

Description: Flowfile is an AI-powered platform designed to streamline and automate repetitive tasks for real estate teams. Its features include document automation, lead qualification, and data management, freeing up valuable time for agents to focus on high-impact activities.
Rating: 95
Reasoning (one sentence): Replacing tedious tasks for real estate agents
ONET Job: Real Estate Agent

Description: ModeWalk is a fashion website catered towards women, offering personalized wardrobe styling services and advice. They aim to help women build a cohesive and stylish wardrobe by providing expert guidance and curated recommendations.
Rating: 34
Reasoning (one sentence): Not going replace the role of a human stylist. It acts more as a tool to assist people with their fashion choices
ONET Job: N/A

Description: AI Mavericks hosts high-profile events centered around AI and business growth, targeting founders, CEOs, and leadership teams seeking to leverage AI for strategic advantage. Their events provide an immersive platform for learning, networking, and gaining practical insights on integrating AI into business strategies.
Rating: 11
Reasoning (one sentence): Not replacing an occupation or task, just hosting events
ONET Job: N/A

**YOUR TURN**
Comapny name: $company
Website: $website
Description: $description
________________________
Rating: --
Reasoning: --
________________________
"""
```

```python
args = [["company","organization name"], ["website","website"], ["description","description_all"]]

async def main(df):
    df = await prompting_class.iterate(df, filter_prompt, args, "automation_rating_llm")
    df.to_csv('../output/df_with_filter.csv', index=False)
    return df

df = asyncio.run(main(df))
```

```python
cols = ["automation_rating", "automation_rating_reasoning"]

def extract_data(text):
    if pd.isnull(text): return {}
    text = re.sub(r'\s+', ' ', text).replace("\n"," ").replace("  "," ").replace("*","").replace("/10","").replace("_","").replace("#","")
    pattern = r"Rating: (\d+)\s+Reasoning:\s+(.*)"

    # Search for the pattern in the text
    match = re.search(pattern, text, re.DOTALL)
    if match:
        automation_rating = match.group(1).strip()
        automation_rating_reasoning = match.group(2).strip()

        result = {
            "automation_rating": automation_rating,
            "automation_rating_reasoning": automation_rating_reasoning
        }
        return result

    else:
        print(text)
        print("FAILURE TO MATCH")
        return {}
        
for col in cols:
    if col in list(df.columns): df=df.drop(columns=[col])
        
results_df = pd.DataFrame(list(df.apply(lambda x: extract_data(x["automation_rating_llm"]),axis=1)))
df = pd.concat([df, results_df],axis=1)
```

```python
df = df.dropna(subset=["automation_rating"],axis=0)
df = df.drop_duplicates(subset=['organization name'], keep='first')
df = df.reset_index(drop=True)
df.to_csv('../output/df_with_filter.csv', index=False)
```

```python
# Load and process the data
df = pd.read_csv("../output/df_with_filter.csv")
df['founded date'] = pd.to_datetime(df['founded date'])
df['year'] = df['founded date'].dt.year
len(df)
```

```python
import plotly.express as px
import plotly.graph_objects as go
import pandas as pd
import numpy as np
from scipy.ndimage import gaussian_filter1d
import plotly.io as pio


grouped = df.groupby('year').agg({'automation_rating':'mean','organization name':'count'}).reset_index()

# Apply Gaussian smoothing
smoothed_ratings = gaussian_filter1d(grouped['automation_rating'], sigma=2)

# Create the figure
fig = go.Figure()

# Add the scatter plot
fig.add_trace(go.Scatter(
    x=grouped["year"], 
    y=grouped["automation_rating"], 
    mode='markers', 
    name='Automation Rating'
))

# Add the smooth line
fig.add_trace(go.Scatter(
    x=grouped["year"], 
    y=smoothed_ratings, 
    mode='lines', 
    name='Smoothed Line',
    line=dict(color='firebrick', width=2)
))

# Update layout with titles and dimensions
fig.update_layout(
    title="Automation Rating of Ecommerce Startups Founded During Year (13,100 startups)",
    xaxis_title="Year Founded",
    yaxis_title="Mean Automation Rating for Year (1-100)",
    width=800,
    height=600
)

# Add the annotation for the caption
fig.add_annotation(
    text="ChatGPT asked: Is the goal of the product of company to be used by a task & occupation in the US workforce?",
    xref="paper", yref="paper",
    x=0.5, y=-0.2,
    showarrow=False,
    font=dict(size=12),
    xanchor='center'
)

# Show the plot
fig.show()
pio.write_image(fig, '../output/automation_rating_over_years.png')
```
